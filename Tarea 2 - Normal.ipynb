# > Función de densidad

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

#### Lista de 40 elementos con valores aleatorios reales entre -10 y +10

```python
def crear_lista():
    lista = []
    for i in range(40):
        lista.append(round(random.uniform(-10, 10),3))
    return lista

elementos = crear_lista()
print(elementos)
```

#### Cálculo de la Media y la Desviación Típica

```python
m = np.mean(elementos, axis = 0) # Media
d = np.std(elementos, axis = 0) # Desviación típica
print("Media: ", m, ", Desviación típica: ", d)
```

#### Histograma

```python
def histograma(lista):
    plt.hist(lista) # Histograma
    plt.show()

histograma(elementos)
```

#### Gráfica con la función de densidad normal

```python
def mostrar_grafica():
    rango = np.arange(-10, 10, 0.01)
    plt.plot(rango, norm.pdf(rango, m, d)) # Distribución normal
    plt.show()

mostrar_grafica()
```
